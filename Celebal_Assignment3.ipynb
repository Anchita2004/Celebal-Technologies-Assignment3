```python
#Import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/IRIS.csv")
```

```python
print(data.head)
```

```python
# Set plot style
sns.set(style="darkgrid")
```

```python
#Bar Chart: Count of each species
plt.figure(figsize=(6,4))
sns.countplot(x='species', data=data, palette='viridis')
plt.title('Count of Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#Line Chart: Sepal Length vs. Sepal Width
plt.figure(figsize=(8,5))
for species in data['species'].unique():
    subset = data[data['species'] == species]
    plt.plot(subset['sepal_length'], subset['sepal_width'], marker='o', linestyle='-', label=species)
plt.title('Sepal Length vs. Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Histogram: Distribution of petal length
plt.figure(figsize=(7,4))
sns.histplot(data['petal_length'], kde=True, bins=20, color='teal')
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.tight_layout()
plt.show()
```

```python
#Pie Chart: Proportion of each species
species_counts = data['species'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Species Proportion in Dataset')
plt.tight_layout()
plt.show()
```

```python
#Box and Whisker Plot: Petal length by species
plt.figure(figsize=(8,5))
sns.boxplot(x='species', y='petal_length', data=data, palette='Set2')
plt.title('Boxplot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.tight_layout()
plt.show()
```
